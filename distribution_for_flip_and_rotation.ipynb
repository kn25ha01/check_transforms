```python
import numpy as np
from PIL import Image
from torchvision import transforms
```

```python
# 水平反転のみ
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    #transforms.RandomVerticalFlip(p=0.5),
    #transforms.RandomRotation(90),
])

image_add = np.zeros([150, 200, 3])
image_sum = np.zeros([150, 200, 3])

for i in range(1000):
    #image_add[, 50, :] += 1
    image_add[45:55, 45:55, :] += 1
    image_add[70:80, 30:40, :] += 1
    image_sum += np.array(transform(Image.fromarray(np.uint8(image_add))))

Image.fromarray(np.uint8(image_sum/image_sum.max()*255))
```
